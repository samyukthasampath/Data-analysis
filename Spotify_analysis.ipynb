```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks=pd.read_csv("/content/drive/MyDrive/Dataset/tracks.csv.zip")
```

```python
df_tracks.head()
```

```python
#check null valus
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
# 10 least popular tracks
sorted_df=df_tracks.sort_values('popularity',ascending=True).head()
sorted_df
```

```python
# 10 most popular tracks
mostpopular_df=df_tracks.sort_values('popularity',ascending=False).head(10)
mostpopular_df
```

```python
df_tracks.describe().transpose()
```

```python
df_tracks.set_index("release_date",inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[16]
```

```python
df_tracks[["artists"]].iloc[45]
```

```python
corr_df=df_tracks.drop(["key",'mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt='.1g',vmin=-1,vmax=1,center=0,cmap='inferno',linewidth=1,linecolor='Black')
heatmap.set_title("Correlation Heat Map Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
print (len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='loudness',x='energy',color='c').set(title="Loudness Vs Energy Correlation ")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='popularity',x='acousticness',color='b').set(title="Popularity Vs Acousticness Correlation ")
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title='Number of songs per year')
```

```python
total_df=df_tracks.duration
fig_dims=(18,7)
fig,ax= plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_df,ax=ax,errwidth=False).set(title="Years vs Duration")
plt.xticks(rotation=90)
```

```python
genre_df=pd.read_csv("/content/drive/MyDrive/Dataset/SpotifyFeatures.csv.zip")
```

```python
genre_df.head()
```

```python
plt.title("Duration of songs in different Genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y='genre',x='duration_ms',data=genre_df)
plt.xlabel('Duration in miliseconds')
plt.ylabel("Genre")
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous=genre_df.sort_values('popularity',ascending=False).head(10)
sns.barplot(y='genre',x='popularity',data=famous).set(title="Top 5 genres by Popularity")
```

```python
#top 5 popular artist
top_five_artist=df_tracks.groupby('artists').count().sort_values(by="name",ascending=False)["name"][:5]
top_five_artist
```

```python
top_five_artist.plot.barh()
plt.show()
```

```python
#top 5 loudest tracks
top_5_loud=df_tracks[['loudness','name']].sort_values(by='loudness',ascending=True)[:5]
top_5_loud
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='loudness',y='name',data=top_5_loud)
plt.show
```

```python
#top 10 instrumentalness tracks
top_10=df_tracks[['instrumentalness','name','artists']].sort_values(by='instrumentalness',ascending=False)[:10]
top_10
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='instrumentalness',y='artists',data=top_10)
plt.show
```

```python
#top 10 danceable tracks
top_10_danceable=df_tracks[['danceability','name','artists']].sort_values(by='danceability',ascending=False)[:10]
top_10_danceable
```

```python
plt.figure(figsize=(12,7))
plt.pie(x='danceability',data=top_10_danceable,autopct='%1.2f%%',labels= top_10_danceable.name)
plt.show
```
